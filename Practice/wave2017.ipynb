```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import sys
sys.path.append('D:\\Coding\\00_Function\\')
from A3_select_data import DataRange
```

```python
def read_excel_data(folder,file_name):
    #year = 2016,2017,2018,2019
    data_path = 'D:/Coding/'
    df   = pd.read_excel(data_path + folder + file_name)
    observations = df.copy()
    observations['DateTime'] = pd.to_datetime(observations['DateTime'])
    print("Import '{}'Done".format(file_name))
    return observations
```

```python
df = read_excel_data('01_Data/02_Wave/','2017.xlsx')
```
